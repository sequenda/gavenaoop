# Introduction to Jupyter Notebooks
Jupyter Notebooks are interactive documents that allow you to write and execute code, visualize data, and add explanatory text all in one place. They are widely used for data science, machine learning, and education.

## Key Features
- **Interactive code execution**
- **Rich text support (Markdown, LaTeX)**
- **Data visualization**
- **Easy sharing and collaboration**

## Basic Structure of a Notebook
A notebook consists of cells. The two main types are:
- **Code cells**: Run code and display output.
- **Markdown cells**: Add formatted text, images, and equations.

````python
# Example: Python code cell
print('Hello, Jupyter!')
````

## Using Markdown Cells
You can write formatted text using Markdown. For example:
- **Bold**
- *Italic*
- Lists
- [Links](https://jupyter.org)

## Visualizing Data
Jupyter supports many visualization libraries. Here is a simple example using matplotlib:

````python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])
plt.title('Simple Plot')
plt.show()
````

## How to Use Jupyter Notebooks in VS Code
Visual Studio Code (VS Code) provides excellent support for Jupyter Notebooks.

### Step 1: Install the Jupyter Extension
1. Open VS Code.
2. Go to the Extensions view by clicking the square icon on the sidebar or pressing `Ctrl+Shift+X`.
3. Search for "Jupyter".
4. Click "Install" on the Jupyter extension by Microsoft.

### Step 2: Create and Run a Notebook
- Click the "New File" button and select "Jupyter Notebook".
- Add code or markdown cells and run them using the play button.

## Tips

- Use keyboard shortcuts for faster editing (e.g., `Shift+Enter` to run a cell).

# Markdown Lecture
Markdown is a lightweight markup language used to format text. In Jupyter Notebooks, markdown cells allow you to create headings, lists, links, images, and even mathematical equations.

## Common Markdown Syntax
- **Headings:** Use `#`, `##`, `###` for different heading levels.
- **Bold:** `**text**` or `__text__`
- *Italic:* `*text*` or `_text_`
- `Code`: Use backticks \`code\`
- Lists: Use `-` or `*` for bullet points
- Numbered lists: Use `1.`, `2.`, etc.
- [Links](https://jupyter.org): `[text](url)`
- Images: `![alt text](image_url)`
- Blockquotes: `> quote`
- Horizontal line: `---`

## Example
```
# Heading 1
## Heading 2
**Bold text**
*Italic text*
- Item 1
- Item 2
[Visit Jupyter](https://jupyter.org)
```

## Mathematical Equations
You can write equations using LaTeX syntax:
- Inline: `$a^2 + b^2 = c^2$`
- Block:
$$
E = mc^2
$$
