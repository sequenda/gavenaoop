## Exercise 1
Check if integer is a prime or not, and display the result.

```python
is_prime = True

while True:
    user_number = input("Number: ")

    if user_number.lower() == 'q':
        print("Bye!")
        exit()
    try:
        n = int(user_number)
        break
    except ValueError:
        print("Not an integer.")

if n <= 3:
    # Check if below 1, which includes 0 which is not a prime.
    is_prime = n > 1
elif n % 2 == 0 or n % 3 == 0:
    is_prime = False
else:
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            is_prime = False
            break
        else:
            i += 6

if is_prime:
    print("Prime.")
else:
    print("Not prime.")

```

## Exercise 2
Compute and display the cube of integers.

```python
array_size = int(input("Enter the size of array: "))
elements = input(f"Enter {array_size} elements, separated with a space: ").split()

if len(elements) != array_size: # mismatch check
    print(f"Warning: {len(elements)} numbers were provided, {array_size} expected.")

arr_1 = []

for i in range(min(array_size, len(elements))): # select which is lower
    num = int(elements[i])
    arr_1.append(num)

print("Cubes:")
for j in arr_1:
    print(j ** 3)
```

## Exercise 3
Generate a hollow square based on the user's integer input side length n.

```python
side_length = int(input("Enter the side length of the square: "))

i = 1
while i <= side_length:
    if i == 1 or i == side_length:
        print("x" * side_length)
    else:
        print("x", end="")
        print(" " * (side_length - 2), end="")
        print("x")
    i += 1 
```

# Exercise 4
Generate an inverted triangle of height n.

```python
while True:
    try:
        triangle_height = int(input("Enter the height of the triangle: "))
        if triangle_height < 1:
            print("Less than 1. Try again.")
            continue
        else:
            break
    except ValueError:
        print("Not an integer.")
        

i = triangle_height

while i >= 1:
    print("*" * i)
    i -= 1
```
